# 1.- EXTRACCION DE DATOS

```python
import pandas as pd

# URL del archivo JSON (versión raw)
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json'

# Cargar el dataset
df = pd.read_json(url)

# Mostrar primeras filas
print("Dimensiones:", df.shape)
df.head()
```

# 2.- TRANSFORMACION (T-TRANSFORM)

```python
# Separar las columnas que contienen diccionarios
customer_data = pd.json_normalize(df['customer'])
phone_data = pd.json_normalize(df['phone'])
internet_data = pd.json_normalize(df['internet'])
account_data = pd.json_normalize(df['account'])

# Unir todo con el dataframe principal (sin las columnas anidadas originales)
df_flat = pd.concat([
    df.drop(columns=['customer', 'phone', 'internet', 'account']),
    customer_data,
    phone_data,
    internet_data,
    account_data
], axis=1)

# Vista previa del nuevo DataFrame "aplanado"
print(df_flat.shape)
df_flat.head()
```

```python
# 1. Convertir 'Charges.Monthly' a numérico
df_flat['Charges.Monthly'] = pd.to_numeric(df_flat['Charges.Monthly'], errors='coerce')

# 2. Crear la nueva columna 'Cuentas_Diarias' dividiendo por 30 días
df_flat['Cuentas_Diarias'] = (df_flat['Charges.Monthly'] / 30).round(2)

# 3. Vista previa
df_flat[['Charges.Monthly', 'Cuentas_Diarias']].head()
```

```python
# Contar cuántos valores no pudieron convertirse y quedaron como NaN
nulos_mensual = df_flat['Charges.Monthly'].isnull().sum()
print(f"Valores no numéricos o faltantes en 'Charges.Monthly': {nulos_mensual}")

# Mostrar algunas filas donde 'Charges.Monthly' es NaN
df_flat[df_flat['Charges.Monthly'].isnull()].head()
```

# 2.5.- ESTANDARIZACION DE DATOS

```python
# Lista de columnas binarias que contienen "Yes"/"No"
columnas_binarias = [
    'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies',
    'PaperlessBilling', 'Churn'
]

# Aplicar conversión a 1/0
for col in columnas_binarias:
    df_flat[col] = df_flat[col].map({'Yes': 1, 'No': 0})
```

```python
df_flat = df_flat.rename(columns={
    'gender': 'Genero',
    'SeniorCitizen': 'AdultoMayor',
    'Partner': 'Pareja',
    'Dependents': 'Dependientes',
    'PhoneService': 'ServicioTelefonico',
    'MultipleLines': 'LineasMultiples',
    'OnlineSecurity': 'SeguridadOnline',
    'OnlineBackup': 'RespaldoOnline',
    'DeviceProtection': 'ProteccionDispositivo',
    'TechSupport': 'SoporteTecnico',
    'StreamingTV': 'TVStreaming',
    'StreamingMovies': 'PeliculasStreaming',
    'Contract': 'TipoContrato',
    'PaperlessBilling': 'FacturacionElectronica',
    'PaymentMethod': 'MetodoPago',
    'Charges.Monthly': 'Facturacion_Mensual',
    'Charges.Total': 'Facturacion_Total',
    'Churn': 'CanceloServicio',
    'Cuentas_Diarias': 'Facturacion_Diaria'
})
```

```python
# Limpiar espacios extra de columnas tipo texto
for col in df_flat.select_dtypes(include='object').columns:
    df_flat[col] = df_flat[col].str.strip()
```

# 3.1.- Cancelacion por genero

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.countplot(x='CanceloServicio', data=df_flat)
plt.title('Distribución de Cancelación de Clientes')
plt.xlabel('¿Canceló el servicio? (0 = No, 1 = Sí)')
plt.ylabel('Cantidad de clientes')
plt.xticks([0, 1], ['No', 'Sí'])
plt.show()
```

```python
# Calcular proporciones
labels = ['Se quedaron', 'Cancelaron']
sizes = df_flat['CanceloServicio'].value_counts(normalize=True).sort_index() * 100

plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'salmon'])
plt.title('Proporción de clientes que cancelaron vs. se quedaron')
plt.axis('equal')
plt.show()
```

# 4.-Extra

Crear una variable de "cantidad de servicios contratados"

```python
# Contar cuántos servicios tiene cada cliente (suma de columnas binarias)
servicios = [
    'ServicioTelefonico', 'LineasMultiples', 'SeguridadOnline',
    'RespaldoOnline', 'ProteccionDispositivo', 'SoporteTecnico',
    'TVStreaming', 'PeliculasStreaming'
]

df_flat['CantidadServicios'] = df_flat[servicios].sum(axis=1)
```
